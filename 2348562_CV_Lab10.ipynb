## **Frame extraction from the video and display the color components of the images**

Implemented frame extraction for two videos:

1st video:

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Loading the video
video_path = "/content/Sowmya C-cv video.mp4"
cam = cv2.VideoCapture(video_path)

# Getting the frame rate of the video
fps = cam.get(cv2.CAP_PROP_FPS)

# Calculating the number of frames to skip for 1 second interval
skip_frames = int(fps)

frame_no = 0

# Reading the frames
while (True):
    ret, frame = cam.read()

    if ret:
        frame_no += 1
        if frame_no % skip_frames == 0:
            # Extracting the color components
            b, g, r = cv2.split(frame)
            print(f"Frame Number: {frame_no}")
            print(f"Frame Shape: {frame.shape}")
            print(f"Color Components are as follows:\nRed - {np.mean(r):.2f}\nGreen - {np.mean(g):.2f}\nBlue - {np.mean(b):.2f}")
            plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            plt.show()
    else:
        break

cam.release()
```

```python
video_path = "/content/Sowmya C-cv video.mp4"

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

ret, frame = cap.read()

if not ret:
    print("Error: Failed to read frame from the video.")
    exit()

frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

blue, green, red = cv2.split(frame_rgb)

plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.imshow(frame_rgb)
plt.title('Original Frame')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(red, cmap='Reds')
plt.title('Red Component')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(green, cmap='Greens')
plt.title('Green Component')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(blue, cmap='Blues')
plt.title('Blue Component')
plt.axis('off')

plt.tight_layout()
plt.show()

cap.release()
```
