Feature Extraction


Loading the images

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('/content/image.jpg',0)
img1=cv2.imread('/content/set-2-q2.tif',0)
img2=cv2.imread('/content/car.png',0)
plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.title("Original Image 1")
plt.imshow(img,"gray")
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Original Image 2")
plt.imshow(img1,"gray")
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Original Image 3")
plt.imshow(img2,"gray")
plt.axis("off")
```

Point Detection

```python

point = np.array([[1, 1, 1],
                [1, -8, 1],
                [1, 1, 1]])
point_img = cv2.filter2D(src=img, ddepth=-1, kernel=point)
point_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=point)
point_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=point)
plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.title("Points Detected in Image 1")
plt.imshow(point_img,"gray")
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Points Detected in Image 2")
plt.imshow(point_img1,"gray")
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Points Detected in Image 3")
plt.imshow(point_img2,"gray")
plt.axis("off")
```

 Line detection

```python

hor_line = np.array([[-1, -1, -1],
                    [2, 2, 2],
                    [-1, -1, -1]])
hor_line_img = cv2.filter2D(src=img, ddepth=-1, kernel=hor_line)
hor_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=hor_line)
hor_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=hor_line)
ver_line = np.array([[-1, 2, -1],
                    [-1, 2, -1],
                    [-1, 2, -1]])
ver_line_img = cv2.filter2D(src=img, ddepth=-1, kernel=ver_line)
ver_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=ver_line)
ver_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=ver_line)
plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.title("Horizontal Line Detected in Image 1")
plt.imshow(hor_line_img,"gray")
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Horizontal Line Detected in Image 2")
plt.imshow(hor_line_img1,"gray")
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Horizontal Line Detected in Image 3")
plt.imshow(hor_line_img2,"gray")
plt.axis("off")
plt.subplot(2,4,1)
plt.title("Vertical Line Detected in Image 1")
plt.imshow(ver_line_img,"gray")
plt.axis("off")
plt.subplot(2,4,2)
plt.title("Vertical Line Detected in Image 2")
plt.imshow(ver_line_img1,"gray")
plt.axis("off")
plt.subplot(2,4,3)
plt.title("Vertical Line Detected in Image 3")
plt.imshow(ver_line_img2,"gray")
plt.axis("off")
```

 Slanting line detection

```python

pos45_line = np.array([[-1, -1, 2],
                    [-1, 2, -1],
                    [2, -1, -1]])
pos45_line_img = cv2.filter2D(src=img, ddepth=-1, kernel=pos45_line)
pos45_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=pos45_line)
pos45_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=pos45_line)


plt.figure(figsize=(15,10))
plt.subplot(1,4,1)
plt.title("+45 Slanting in Image 1")
plt.imshow(pos45_line_img,"gray")
plt.axis("off")
plt.subplot(1,4,2)
plt.title("+45 Slanting in Image 2")
plt.imshow(pos45_line_img1,"gray")
plt.axis("off")
plt.subplot(1,4,3)
plt.title("+45 Slanting in Image 3")
plt.imshow(pos45_line_img2,"gray")
plt.axis("off")
```

```python
neg45_line = np.array([[2, -1, -1],
                    [-1, 2, -1],
                    [-1, -1, 2]])
neg45_line_img = cv2.filter2D(src=img, ddepth=-1, kernel=neg45_line)
neg45_line_img1 = cv2.filter2D(src=img1, ddepth=-1, kernel=neg45_line)
neg45_line_img2 = cv2.filter2D(src=img2, ddepth=-1, kernel=neg45_line)

plt.figure(figsize=(15,10))
plt.subplot(2,4,1)
plt.title("-45 Slanting in Image 1")
plt.imshow(neg45_line_img,"gray")
plt.axis("off")
plt.subplot(2,4,2)
plt.title("-45 Slanting in Image 2")
plt.imshow(neg45_line_img1,"gray")
plt.axis("off")
plt.subplot(2,4,3)
plt.title("-45 Slanting in Image 3")
plt.imshow(neg45_line_img2,"gray")
plt.axis("off")

```
