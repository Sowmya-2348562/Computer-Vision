### **AVERAGE FILTER ( NOISE FILTERING )**

Built-in function

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Load an image
image = cv2.imread('/content/noisee.png')

# Apply average filter (noise filtering)
average_filtered_image = cv2.blur(image, (5, 5))

# Display the results
cv2_imshow(image)
cv2_imshow(average_filtered_image)

```

User-defined function

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def mean_filter(image, kernel_size):
    output_image = np.zeros_like(image)

    for i in range(kernel_size // 2, image.shape[0] - kernel_size // 2):
        for j in range(kernel_size // 2, image.shape[1] - kernel_size // 2):
            neighborhood = image[i - kernel_size // 2:i + kernel_size // 2 + 1,
                                 j - kernel_size // 2:j + kernel_size // 2 + 1]
            mean_value = np.mean(neighborhood)
            output_image[i, j] = mean_value

    return output_image

# Load an image
image = cv2.imread('/content/noisee.png', cv2.IMREAD_GRAYSCALE)

# Define the size of the kernel (neighborhood)
kernel_size = 3

# Apply mean filtering using the user-defined function
output_image = mean_filter(image, kernel_size)

# Display the results
cv2_imshow(image)
cv2_imshow(output_image)
```

### **GAUSSIAN BLUR**

Built-in function

```python
# Load an image
image = cv2.imread('/content/gaussian.jpg')

# Apply Gaussian blur
gaussian_blurred_image = cv2.GaussianBlur(image, (5, 5), sigmaX=0)

# Display the results
cv2_imshow(image)
cv2_imshow(gaussian_blurred_image)

```

User-defined function

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def gaussian_blur(image, kernel_size):
    # Create Gaussian kernel
    kernel = cv2.getGaussianKernel(kernel_size, 0)
    kernel = np.outer(kernel, kernel.transpose())

    # Apply convolution
    blurred_image = cv2.filter2D(image, -1, kernel)

    return blurred_image

# Load an image
image = cv2.imread('/content/gaussian.jpg')

# Apply Gaussian blur using the user-defined function
blurred_image_def = gaussian_blur(image, 5)

# Display the results
cv2_imshow(image)
cv2_imshow(blurred_image_def)
```

## **SHARPENING FILTER AND LAPLACIAN FILTER**

Built-in function

```python
# Load an image
image = cv2.imread('/content/moon1.png')


# Define the sharpening kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]], dtype=np.float32)

# Apply sharpening filter (Laplacian)
sharpened_image = cv2.filter2D(image, -1, kernel)

# Display the results
cv2_imshow(image)
cv2_imshow(sharpened_image)
```

Built-in function

```python
# Load an image
image = cv2.imread('/content/moon1.png')

# Apply Laplacian filter
laplacian_image = cv2.Laplacian(image, cv2.CV_64F)

# Convert the Laplacian image to uint8 format
laplacian_image = cv2.convertScaleAbs(laplacian_image)

# Display the results
cv2_imshow(image)
cv2_imshow(laplacian_image)
```

User-defined function

```python
def sharpening_filter(image):
    # Define the sharpening kernel
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]])

    # Apply convolution
    sharpened_image = cv2.filter2D(image, -1, kernel)

    return sharpened_image

# Load an image
image = cv2.imread('/content/moon1.png')

# Apply sharpening filter using the user-defined function
sharpened_image_userdefined = sharpening_filter(image)

# Display the results
cv2_imshow(image)
cv2_imshow(sharpened_image_userdefined)
```

User-defined function

```python
def laplacian_filter(image):
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Define the Laplacian kernel
    kernel = np.array([[0, -1, 0],
                       [-1, 4, -1],
                       [0, -1, 0]], dtype=np.float32)

    # Apply convolution using the Laplacian kernel
    laplacian_image = cv2.filter2D(gray_image, -1, kernel)

    return laplacian_image

# Load an image
image = cv2.imread('/content/moon1.png')

# Apply Laplacian filter using the user-defined function
laplacian_image_userdefined = laplacian_filter(image)

# Display the results
cv2_imshow(image)
cv2_imshow(laplacian_image_userdefined)
```
