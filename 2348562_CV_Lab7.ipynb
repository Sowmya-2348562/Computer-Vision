# De-noising

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/content/pls.jpg',0)
plt.imshow(img,"gray")
```

Adding Salt and Pepper noise

```python
import random
row , col = img.shape
number_of_pixels = random.randint(300, 10000)
for i in range(number_of_pixels):
    y_coord=random.randint(0, row - 1)
    x_coord=random.randint(0, col - 1)
    img[y_coord][x_coord] = 255
number_of_pixels = random.randint(300 , 10000)
for i in range(number_of_pixels):
    y_coord=random.randint(0, row - 1)
    x_coord=random.randint(0, col - 1)
    img[y_coord][x_coord] = 0
plt.imshow(img,"gray")
```

de-noising using min filter

```python
from scipy.ndimage import minimum_filter
min=minimum_filter(img, size=3)
plt.imshow(min,"gray")
```

de-noising using max filter

```python
from scipy.ndimage import maximum_filter
max=maximum_filter(img, size=3)
plt.imshow(max,"gray")
```

de-noising using median filter

```python
from scipy.ndimage import median_filter
mid=median_filter(img, size=3)
plt.imshow(mid,"gray")
```



Apply a median filter to the image to replace each pixel value with the median value in its local neighborhood.
This filter is effective in removing salt-and-pepper noise while preserving edges in the image.
Min Filter:

Apply a min filter to the image to replace each pixel value with the minimum value in its local neighborhood.
The min filter is useful for eliminating noise by smoothing out small variations in pixel intensities.
Max Filter:

Apply a max filter to the image to replace each pixel value with the maximum value in its local neighborhood.
The max filter is helpful in reducing noise by enhancing edges and details in the image.

De-noising an image using median filtering involves replacing each pixel's intensity value by the median value of the intensities in the neighborhood of that pixel. The min and max filters can also be used for de-noising by replacing each pixel's intensity with the minimum or maximum value in its neighborhood.
